#Dictionary:

A dictionary in Python is an unordered collection of key-value pairs. Each key is unique within the dictionary and is used to access its corresponding value. Dictionaries are highly useful for storing and retrieving data based on specific identifiers (keys). They are defined using curly braces {} with key-value pairs separated by colons.

```python
#Creating a Dictionary:
#You can create a dictionary by specifying key-value pairs within curly braces {}:
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
my_dict
```

```python
#Accessing Values:
#You can access the value associated with a specific key:
name = my_dict["name"]
name
```

```python
#Adding or Modifying Key-Value Pairs:
#You can add or modify key-value pairs using assignment:
my_dict["occupation"] = "Engineer"
my_dict
```

```python
#Removing Key-Value Pairs:

#del: Deletes a specific key-value pair.
#pop(): Removes and returns the value associated with a key.

del my_dict["age"]
my_dict
```

```python
popped_value = my_dict.pop("city")
popped_value
```

```python
my_dict
```

```python
#Checking Key Existence:
#The in keyword is used to check if a key exists in the dictionary:
if "name" in my_dict:
    print("Name exists!")
```

```python
#Keys, Values, Items:

#keys(): Returns a list of all keys.
#values(): Returns a list of all values.
#items(): Returns a list of key-value tuples.
```

```python
all_keys = my_dict.keys()
all_values = my_dict.values()
all_items = my_dict.items()
```

```python
all_keys
```

```python
all_values
```

```python
all_items
```

```python
#Copying a Dictionary:
#The copy() method creates a shallow copy of the dictionary:
new_dict = my_dict.copy()
new_dict
```

```python
#Clearing a Dictionary:
#The clear() method empties the dictionary:
my_dict.clear()
my_dict
```

```python
#Getting a Value with Default:
#The get() method returns the value for a key, or a default value if the key doesn't exist:
occupation = my_dict.get("occupation")
occupation
```

```python
#Updating a Dictionary:
#The update() method adds key-value pairs from another dictionary or iterable:
my_dict.update({"age": 25, "city": "London"})
my_dict
```

```python
#Dictionary Size:
#The len() function returns the number of key-value pairs in the dictionary:
dict_size = len(my_dict)
dict_size
```

```python
#Converting Lists to Dictionaries:
#You can convert a list of tuples into a dictionary using the dict() function:
list_of_tuples = [("name", "Alice"), ("age", 30)]
converted_dict = dict(list_of_tuples)
converted_dict
```
