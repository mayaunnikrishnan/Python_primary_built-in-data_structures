### **Type conversion methods within the same data types in Python:**



```python
#Integer:

#Float to Integer: int()

my_float = 3.14
converted_integer = int(my_float)
converted_integer
```

```python
#String to Integer: int()
my_string = "123"
converted_integer = int(my_string)
converted_integer
```

```python
#Float:

#Integer to Float: float()
my_integer = 5
converted_float = float(my_integer)
converted_float
```

```python
#String to Float: float()
my_string = "3.14"
converted_float = float(my_string)
converted_float
```

```python
#String:

#Integer to String: str()
my_integer = 42
converted_string = str(my_integer)
converted_string
```

```python
#Float to String: str()
my_float = 3.14
converted_string = str(my_float)
converted_string
```

```python
#Boolean:

#Integer to Boolean: bool()
my_integer = 0
converted_boolean = bool(my_integer)
converted_boolean
```

```python
#String to Boolean: bool()
my_string = "True"
converted_boolean = bool(my_string)
converted_boolean
```

### **Type conversion methods within different data structures in Python:**

```python
#List:

#Tuple to List: list()
my_tuple = (1, 2, 3)
converted_list = list(my_tuple)
converted_list
```

```python
#String to List: list()
my_string = "hello"
converted_list = list(my_string)
converted_list
```

```python
#Set to List: list()
my_set = {1, 2, 3}
converted_list = list(my_set)
converted_list
```

```python
#Tuple:

#List to Tuple: tuple()
my_list = [1, 2, 3]
converted_tuple = tuple(my_list)
converted_tuple
```

```python
#String to Tuple: tuple()
my_string = "hello"
converted_tuple = tuple(my_string)
converted_tuple
```

```python
#Set to Tuple: tuple()
my_set = {1, 2, 3}
converted_tuple = tuple(my_set)
converted_tuple
```

```python
#Set:

#List to Set: set()
my_list = [1, 2, 3]
converted_set = set(my_list)
converted_set
```

```python
#Tuple to Set: set()
my_tuple = (1, 2, 3)
converted_set = set(my_tuple)
converted_set
```

```python
#String to Set: set()
my_string = "hello"
converted_set = set(my_string)
converted_set
```

```python
 #String:

#List to String: join()
my_list = ["h", "e", "l", "l", "o"]
converted_string = "".join(my_list)
converted_string
```

```python
#Tuple to String: join()
my_tuple = ("h", "e", "l", "l", "o")
converted_string = "".join(my_tuple)
converted_string
```

```python
#Set to String: join()
my_set = {"h", "e", "l", "l", "o"}
converted_string = "".join(my_set)
converted_string
```
