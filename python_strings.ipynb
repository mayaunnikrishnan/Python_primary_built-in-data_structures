#string

In Python, a string is a sequence of characters. Strings are used to store text data and are one of the most commonly used data types. They are defined by enclosing characters within single quotes (' ') or double quotes (" "). For example:



```python
string1 = 'Hello, World!'
string2 = "Python is fun!"
```

##Common String Methods

```python
s = "Hello, World!"
length = len(s)
print(length)  # Output: 13
```

```python
s = "Hello, World!"
lower_s = s.lower()
print(lower_s)  # Output: "hello, world!"
```

```python
s = "Hello, World!"
upper_s = s.upper()
print(upper_s)  # Output: "HELLO, WORLD!"
```

```python
s = "   Hello, World!   "
stripped_s = s.strip()
print(stripped_s)  # Output: "Hello, World!"
```

```python
s = "Hello, World!"
replaced_s = s.replace("World", "Python")
print(replaced_s)  # Output: "Hello, Python!"
```

```python
s = "Hello, World!"
split_s = s.split(", ")
print(split_s)  # Output: ["Hello", "World!"]
```

```python
words = ["Hello", "World"]
joined_s = ", ".join(words)
print(joined_s)  # Output: "Hello, World"
```

```python
s = "Hello, World!"
index = s.find("World")
print(index)  # Output: 7
```

```python
s = "Hello, World!"
result = s.startswith("Hello")
print(result)  # Output: True
```

```python
s = "Hello, World!"
result = s.endswith("World!")
print(result)  # Output: True
```

##String Check Methods

```python
s = "12345"
print(s.isnumeric())  # Output: True
```

```python
s = "12345"
print(s.isdigit())  # Output: True
```

```python
s = "12345"
print(s.isdecimal())  # Output: True
```

```python
s = "Hello"
print(s.isalpha())  # Output: True
```

```python
s = "Hello123"
print(s.isalnum())  # Output: True
```

```python
s = "   "
print(s.isspace())  # Output: True
```

```python
s = "hello"
print(s.islower())  # Output: True
```

```python
s = "HELLO"
print(s.isupper())  # Output: True
```

```python
s = "Hello World"
print(s.istitle())  # Output: True
```

```python
s = "variable1"
print(s.isidentifier())  # Output: True
```
